```python
def myplot(playerlist): 
    for name in playerlist:
        plt.plot(Games[Pdict[name]],color= 'Black', ls = '--',marker='s',ms =7,label=Players[Pdict[name]])
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,9)),Seasons, rotation = 'vertical')
```

```python
# Fix up the Inputs:
def myplot(data,playerlist= Players): 
    Col = {"KobeBryant":"Black","JoeJohnson":"Brown","LeBronJames":"Red","CarmeloAnthony":"Orange","DwightHoward":"Yellow","ChrisBosh":"Green","ChrisPaul":"Pink","KevinDurant":"Magenta","DerrickRose":"Cyan","DwayneWade":"Purple"}
    Mrk = {"KobeBryant":"s","JoeJohnson":"p","LeBronJames":"s","CarmeloAnthony":"o","DwightHoward":"d","ChrisBosh":"^","ChrisPaul":"p","KevinDurant":"s","DerrickRose":"o","DwayneWade":"d"}
    for name in playerlist:
        plt.plot(data[Pdict[name]],color= Col[name], ls = '--',marker=Mrk[name],ms =7,label=Players[Pdict[name]])
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons, rotation = 'vertical')
```
